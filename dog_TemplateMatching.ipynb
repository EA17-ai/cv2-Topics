```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dog=cv2.imread("img_1.png")
dog=cv2.cvtColor(dog,cv2.COLOR_BGR2RGB)
plt.imshow(dog)
```

```python
dogface=cv2.imread("dog-face.png")
dogface=cv2.cvtColor(dogface,cv2.COLOR_BGR2RGB)
plt.imshow(dogface)
```

```python
res=cv2.matchTemplate(dog,dogface,cv2.TM_CCOEFF)
```

```python
plt.imshow(res)
```

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
```

```python
top_left=max_loc
```

```python
length,width,content=dogface.shape
```

```python
bottom_right=(top_left[0]+width,top_left[1]+length)
```

```python
found=cv2.rectangle(dog,top_left,bottom_right,(255,0,0),10)
```

```python
plt.imshow(found)
```
